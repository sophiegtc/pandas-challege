```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df.head()

```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
Columns=["Number of Unique Items","Average Price","Number of Purchases","Total Revenue"]
Number_of_Unique_Items=len(purchase_data_df["Item Name"].unique())
Average_Price=purchase_data_df["Price"].mean()
Number_of_Purchase=len(purchase_data_df["SN"])
Total_Revenue=purchase_data_df["Price"].sum()

#Average_Price=Average_Price.map("${:,.2f}".format)
#Total_Revenue=Total_Revenue.map("${:,.2f}".format)

summary_df=pd.DataFrame({"Number of Unique Items":[Number_of_Unique_Items],"Average Price":[Average_Price],"Number of Purchases":[Number_of_Purchase],"Total Revenue":[Total_Revenue]})
summary_df["Average Price"]=summary_df["Average Price"].map("${:,.2f}".format)
summary_df["Total Revenue"]=summary_df["Total Revenue"].map("${:,.2f}".format)
summary_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins=[0,9,14,19,24,29,34,39,200]
bins_labels=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
Total_players_df=purchase_data_df.drop_duplicates(["SN"])
pd.cut(Total_players_df["Age"],bins,labels=bins_labels, right=False)
Total_players_df["Age Group"]=pd.cut(Total_players_df["Age"],bins,labels=bins_labels)
Total_players_df.head()

unique_count_df=Total_players_df.drop_duplicates(["SN"])
grouped_unique_count=unique_count_df.groupby(['Age Group'])
Total_count=grouped_unique_count.count()
Total_count

#Total_count["Percentage"] =Total_count["Purchase ID"] / Total_count["Purchase ID"].sum()
Percentage=Total_count["Purchase ID"] / Total_count["Purchase ID"].sum() *100
Percentage

summary_data_df=pd.DataFrame({"Total Count":Total_count['SN'],"Percentage of Player":Percentage})
summary_data_df["Percentage of Player"]=summary_data_df["Percentage of Player"].round(2).astype(str) + '%'
                             
summary_data_df
```

```python

pd.cut(purchase_data_df["Age"],bins,labels=bins_labels, right=False)
purchase_data_df["Age Group"]=pd.cut(purchase_data_df["Age"],bins,labels=bins_labels)
purchase_data_df.head()

purchase_age_df =purchase_data_df.groupby(['Age Group'])
purchase_age_count=purchase_age_df.count()
purchase_age_count

grouped_purchase_price=purchase_age_df["Price"].sum()
grouped_purchase_price

average_purchase_price=grouped_purchase_price/purchase_age_df["Age Group"].count()
average_purchase_price


purchase_perperson_df=purchase_data_df.drop_duplicates(["SN"])
players_count_age=purchase_perperson_df["Age Group"].value_counts()
Ag_purchase_person=grouped_purchase_price/players_count_age
Ag_purchase_person

summary_purchase_df=pd.DataFrame({"Purchase Count":purchase_age_count['SN'],"Average Purchase Price":average_purchase_price,
                              "Total Purchase Value":grouped_purchase_price, "Avg Total Purchase per Person":
                                 Ag_purchase_person})

summary_purchase_df["Average Purchase Price"]=summary_purchase_df["Average Purchase Price"].map("${:,.2f}".format)
summary_purchase_df["Total Purchase Value"]=summary_purchase_df["Total Purchase Value"].map("${:,.2f}".format)
summary_purchase_df["Avg Total Purchase per Person"]=summary_purchase_df["Avg Total Purchase per Person"].map("${:,.2f}".format)

summary_purchase_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
grouped_items_df=purchase_data_df.groupby("Item Name")
grouped_items_df.head()

items_count=grouped_items_df["Item Name"].count()
items_count

Total_item_purchase=grouped_items_df["Price"].sum()
Total_item_purchase

Avg_item_price=Total_item_purchase/items_count
Avg_item_price

summary_popular_df=pd.DataFrame({"Purchase Count":items_count,"Item Price":Avg_item_price,
                              "Total Purchase Value":Total_item_purchase})
summary_popular_df

summary_popular_df = summary_popular_df.sort_values(by=["Purchase Count"], ascending=False)

summary_popular_df["Item Price"]=summary_popular_df["Item Price"].map("${:,.2f}".format)
summary_popular_df["Total Purchase Value"]=summary_popular_df["Total Purchase Value"].map("${:,.2f}".format)

summary_popular_df
```
